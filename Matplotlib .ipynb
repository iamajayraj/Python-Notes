# Why do we use plt.plot() if we have plt.scatter() ? 
  
  for plotting scatter plot of a large dataset, the plt.plot() function is relatively faster than plt.scatter().

```python
#colour and label
df=pd.read_csv(r"E:\Datasets\session 23\batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman']) #this function is used  to rename the x-axis values here

plt.show()
```

```python
#The problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
```

```python
#The solution
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')


plt.xticks(rotation='vertical') #here the x-axis values are rotated using rotation parameter

plt.show()
```

```python
#stacked bar chart

plt.bar(df['batsman'],df['2017'])
plt.bar(df['batsman'],df['2016'],bottom=df['2017'])
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']))
```

# Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df=pd.read_csv("E:\Datasets\session 23\gayle-175.csv")
```

```python
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
#explode
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
