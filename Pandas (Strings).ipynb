```python
# Common Functions
# lower/upper/capitalize/title
titanic['Name'].str.lower()
```

```python
titanic['Name'].str.capitalize()
```

```python
#strip-- removes all leading aand trailing spaces
titanic['Name'].str.strip()
```

```python
#find passenger having longest name
titanic['Name'][titanic['Name'].str.len() == 82].values[0]
```

```python
titanic['Last_name']=titanic['Name'].str.split(',').str.get(0)
```

```python
titanic['Title']=titanic['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1).str.get(0)
```

```python
titanic['First_name']=titanic['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1).str.get(1)
```

```python
titanic
```

```python
titanic['Title'].value_counts()
```

```python
titanic['Title']=titanic['Title'].str.replace('Ms.','Miss.')
```

```python
titanic['Title'].value_counts()
```

```python
titanic[titanic['First_name'].str.startswith('A')]
```

```python
# applying regex
# contains
# search john -> both case
titanic[titanic['First_name'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
titanic[titanic['Last_name'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
