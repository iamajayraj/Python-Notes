# Must Read for Subplots
  Here what actually is happening is that when you create fig,ax=plt.subplots(nrows=2,ncols=2). Then, ax  is an 2D array of 4 elements and each element represents a graph in a 2x2 grid of graphs. fig is a single space for plotting graphs. ax is the axes for each graph. To access each graph, 2D array indexing is used, its methods are also acccessed and used using same methods.

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(12,10))
```

```python
fig
```

```python
ax
```

```python
ax[(0,0)].scatter(batters['avg'],batters['runs'])
ax[(0,0)].set_title('Average VS Runs')
ax[(0,0)].set_ylabel("Runs")
```

```python
ax[1,0].scatter(batters['avg'],batters['strike_rate'])
ax[1,0].set_title('Average VS Strike Rate')
ax[1,0].set_xlabel('Average')
ax[1,0].set_ylabel('Strike Rate')
```

# 3D Surface Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x
```

```python
y
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
zz=xx**2+yy**2
```

```python
zz
```

```python
zz.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

plot=ax.plot_surface(xx,yy,zz,cmap='viridis')
plt.colorbar(plot)

#all the values like  the title , xlabel , ylabel can be added here.
```

```python
zz=np.sin(xx)+np.cos(yy)

fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

plot=ax.plot_surface(xx,yy,zz,cmap='viridis')
plt.colorbar(plot)
```
